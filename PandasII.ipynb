```python
def strNumbers(x):
    if "M" in x:
        string = x.strip("M")
        number = float(string)
        number = number * 1000000
        
    else:   
        string = x.strip("K")
        number = float(string)
        number = number * 1000
    return number
```

```python
strNumbers('169.34 K')
```

```python
df["FunVolume"] = df["Volume"].apply(strNumbers)
```
